```python
Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по правилам:

оценка 2 и ниже — низкий рейтинг;
оценка 4 и ниже — средний рейтинг;
оценка 4.5 и 5 — высокий рейтинг.
Результат классификации запишите в столбец class.
```

```python
import pandas as pd
ratings = pd.read_csv('Files for studies/ratings2.csv')
movies = pd.read_csv('Files for studies/movies.csv')
rated_movies = ratings.merge(movies, on =['movieId'])
rated_movies.head()
```

```python
def count_rating(row):
    if row['rating']>=2 and row['rating']<=4:
        return 'Средний рейтинг'
    elif row['rating']>=4 and row['rating']<=5:
        return 'Высокий рейтинг'
    else:
        return 'Низкий рейтинг'
```

```python
rated_movies['class'] = rated_movies.apply(count_rating,axis=1)
```

Задание 2
Используйте файл keywords.csv.

Нужно написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определённому региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

```python
import pandas as pd
keywords = pd.read_csv('Files for studies/keywords.csv')
keywords.head(10)
```

```python
def city_region(words):
    geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
    for k,v in geo_data.items():
        for i in v:
            if k == 'Центр' and i in words['keyword']:
                return 'Центр'
    for k,v in geo_data.items():
        for i in v:
            if k == 'Северо-Запад' and i in words['keyword']:
                return 'Северо-Запад'
    for k,v in geo_data.items():
        for i in v:
            if k == 'Дальний Восток' and i in words['keyword']:
                return 'Дальний Восток'
    else:
        return 'undefined'
```

```python
keywords['region'] = keywords.apply(city_region,axis=1)
```
